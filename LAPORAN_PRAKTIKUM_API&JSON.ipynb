PRAKTIK 1: Request ke API

```python
import requests

url = "https://api.agify.io?name=budi"
response = requests.get(url)

print(response)
print(response.text)
```

PRAKTIK 2: Menggunakan JSON

```python
import requests

url = "https://api.agify.io?name=budi"
response = requests.get(url)

data = response.json()

print(data)
print(data["name"])
print(data["age"])
```

PRAKTIK 3: Input dari User

```python
import requests

nama = input("Masukkan nama: ")
url = f"https://api.agify.io?name={nama}"

response = requests.get(url)
data = response.json()

print(f"Halo, {data['name']}!")
print(f"Perkiraan umur: {data['age']} tahun")
```

PRAKTIK 4: API Quotes

```python
import requests

url = "https://quotes.liupurnomo.com/api/quotes/random"
data = requests.get(url, verify=False).json()

print(data["data"]["text"])
print("-", data["data"]["author"])
```

PRAKTIK 5: Looping API

```python
import requests

for i in range(3):
    data = requests.get("https://quotes.liupurnomo.com/api/quotes?", verify=False).json()

    print(f"--- Quote ke-{i+1} ---")
    print(data["data"][i+1]["text"])
    print(f"- {data["data"][i+1]['author']}")
    print()
```
